```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = Axes3D(fig)

x=np.arange(-5,5,0.5)
y=np.arange(-5,5,0.5)

x,y = np.meshgrid(x,y)

r = np.sqrt(x**2,y**2)

z = np.sin(r)

ax.plot_surface(x,y,z,rstride=1,cstride=1,cmap=plt.get_cmap('rainbow'))

ax.contourf(x,y,z,zdir='x',offset=-5,cmap='rainbow')

ax.set_zlim(-2,2)

plt.show()
```
