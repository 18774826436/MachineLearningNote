```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-1,1,50)

y1 = 2*x + 1
y2 = x**2

# plt.figure()

# plt.plot(x,y1,color='red',linewidth=2.0,linestyle='--')

plt.figure(num=4,figsize=(10,10))
l1 = plt.plot(x,y1,color='red',linewidth=2.0,linestyle='--',label='up')
l2 = plt.plot(x,y2,label='down')
plt.xlim((-1,2))
plt.ylim((-2,3))
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend(labels=['aaaaa','bbbbb'],loc='best')


#坐标标注
new_ticks = np.linspace(-1,2,5)
print(new_ticks)
plt.xticks(new_ticks)

plt.yticks([-2,-1.8,-1,1.22,3],
          [r'$really\ bad$',r'$bad \alpha$','nornal','good','really good'])

#设置轴的属性
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))


plt.show()
```
